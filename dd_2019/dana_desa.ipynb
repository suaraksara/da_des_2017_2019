```python
import pandas as pd
from pandas.io.json import json_normalize
```

```python
from os import listdir
%autosave 900
```

```python
file = [i for i in listdir() if i.endswith('.json')]

daftar_pd = [pd.read_json(i) for i in file]

daftar_data = [json_normalize(i['data']) for i in daftar_pd]

data_gab = pd.concat(daftar_data)

data_gab
```

```python
data_gab.to_csv('dana_desa_2019.csv')
```
